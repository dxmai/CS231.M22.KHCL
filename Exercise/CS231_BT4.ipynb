# Hàm tính Histogram

```python
import cv2 as cv
```

```python
def TinhHist(pathfilename):
  img = cv.imread(pathfilename,0)
  hist = cv.calcHist([img],[0],None,[256],[0,256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size
  return hist
```

# Hàm tính HOG

```python
from skimage import feature
```

```python
def TinhHOG(pathfilename):
  img = cv.imread(pathfilename,0)
  img =cv.resize(img, (128, 256))
  (hog, hog_image) = feature.hog(img, orientations=9,
  pixels_per_cell=(8, 8), cells_per_block=(2, 2),
  block_norm='L2-Hys', visualize=True, transform_sqrt=True)
  return hog
```

# Hàm Extract Features

```python
def ExtractFeatures(pathfilename, option=0):
  if option == 0:
    return TinhHist(pathfilename)
  return TinhHOG(pathfilename)
```

# Hàm đọc dữ liệu

```python
import numpy as np
```

```python
def ReadData(path_to_files, option=0):
  features, labels = [], []
  for file_name in os.listdir(path_to_files):
    imagePath = os.path.join(path_to_files, file_name)
    label = imagePath.split(os.path.sep)[-1].split(".")[0]
    labels.append(label)
    feature = ExtractFeatures(imagePath, option)
    features.append(feature)
  return np.array(features), np.array(labels)
```

# Hàm k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def K_NN(x_train, y_train, x_test, y_test, n):
  model = KNeighborsClassifier(n_neighbors=n)
  model.fit(x_train, y_train)
  acc = model.score(x_test, y_test)
  return acc
```

# Hàm Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
def LogisticRegressionModel(x_train, y_train, x_test, y_test):
  model = LogisticRegression()
  model.fit(x_train, y_train)
  acc = model.score(x_test, y_test)
  return acc
```

# Áp dụng cho Đặc trưng Histogram

```python
features_train, labels_train = ReadData(dir_train)
```

```python
features_test, labels_test = ReadData(dir_test)
```

## Điều chỉnh lại giá trị của shape

```python
nsamples, nx, ny = features_train.shape
d2_features_train = features_train.reshape((nsamples,nx*ny))
```

```python
nsamples, nx, ny = features_test.shape
d2_features_test = features_test.reshape((nsamples,nx*ny))
```

Ref: https://www.codegrepper.com/code-examples/python/ValueError%3A+Found+array+with+dim+3.+Estimator+expected+%3C%3D+2.+sklearn

## k-NN

```python
neighbors = [1, 2, 3]
for each in neighbors:
  print("Accuracy when k = {}: {}".format(each, K_NN(d2_features_train, labels_train, d2_features_test, labels_test, each)))
```

## Logistic Regression

```python
print("Accuracy when using Logistic Regression for Histogram: ", LogisticRegressionModel(d2_features_train, labels_train, d2_features_test, labels_test))
```
